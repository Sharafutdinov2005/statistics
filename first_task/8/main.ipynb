```python
import numpy as np
import pandas as pd
from scipy.stats import chi2
```

```python
# data init
DATA = pd.DataFrame({
        "First group": {
            "Small": 25,
            "Exact": 50,
            "Big": 25
        },
        "Second group": {
            "Small": 52,
            "Exact": 41,
            "Big": 7
        }
    })

n = np.sum(DATA)

DATA
```

```python
# sample probabilities calculation
group_probabilities = DATA.sum() / n
size_probabilities = DATA.sum(axis=1) / n

random_vector_probabilities = pd.DataFrame(
    np.outer(size_probabilities.values, group_probabilities.values),
    index=size_probabilities.index,
    columns=group_probabilities.index,
)
random_vector_probabilities
```

```python
delta = np.sum(
    (DATA - random_vector_probabilities * n) ** 2 /
    (random_vector_probabilities * n)
)

(DATA - random_vector_probabilities * n) ** 2 / (random_vector_probabilities * n)
```

```python
# p-value calculation
chi2.sf(delta,  df=(2 - 1) * (3 - 1))  # 1 - F(x)
```
