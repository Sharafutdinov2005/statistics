```python
import numpy as np
from scipy.optimize import minimize_scalar
from scipy.stats import chi2, poisson
```

```python
# data init
DATA = np.array(
    [0] * 109 +
    [1] * 65 +
    [2] * 22 +
    [3] * 3 +
    [4] * 1
)
```

```python
# maximum group likehood function
def neg_log_L(theta):
    return -(
        65 * np.log(theta) -
        174 * theta +
        26 * np.log(
            np.sum(poisson.pmf(np.arange(2, 5, 1), mu=theta))
        )
    )
```

```python
# maximum group likehood estimaion
tilde_theta = minimize_scalar(neg_log_L, bounds=(0.0, 50.0)).x  # set some large upper bound for theta
tilde_theta
```

```python
# delta estimation
tilde_delta = (
    (np.sum(DATA==0) - DATA.size * poisson.pmf(0, mu=tilde_theta)) ** 2 /
    (DATA.size * poisson.pmf(0, mu=tilde_theta)) +
    (np.sum(DATA==1) - DATA.size * poisson.pmf(1, mu=tilde_theta)) ** 2 /
    (DATA.size * poisson.pmf(1, mu=tilde_theta)) +
    (np.sum(DATA>=2) - DATA.size * np.sum(poisson.pmf(np.arange(2, 5, 1), mu=tilde_theta))) ** 2 /
    (DATA.size * np.sum(poisson.pmf(np.arange(2, 5, 1), mu=tilde_theta)))
)
tilde_delta
```

```python
# p-value calculation, sf()
group_number = 3  # 0, 1, 234
theta_dim = 1     # theta in R^1
chi2.sf(tilde_delta, df=group_number-theta_dim-1)
```
