```python
import numpy as np
import pandas as pd
from scipy.stats import chi2
```

```python
# data init
DATA = pd.DataFrame({
    "First group": {
        2: 33,
        3: 43,
        4: 80,
        5: 144
    },
    "Second group": {
        2: 39,
        3: 35,
        4: 72,
        5: 154
    }
})

N = np.sum(DATA)

DATA
```

```python
# sample probabilities calculation
probabilities = DATA.sum(axis=1) / N
probabilities
```

```python
delta = np.sum(
    (DATA.T - probabilities * 300) ** 2 /
    (probabilities * 300)
)

(DATA.T - probabilities * 300) ** 2 / (probabilities * 300)
```

```python
# p-value calculation
chi2.sf(delta,  df=(4 - 1) * (2 - 1))  # 1 - F(x)
```
